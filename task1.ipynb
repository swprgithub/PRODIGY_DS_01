```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Load the CSV file
file_path = 'C:/Users/LENOVO/Downloads/world_population.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Create a histogram for the 2022 population
plt.figure(figsize=(10, 6))
plt.hist(df['2022 Population'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Population in 2022')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```

```python
# Create a bar chart for the number of countries in each continent
continent_counts = df['Continent'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
continent_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Number of Countries in Each Continent')
plt.xlabel('Continents')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
